## Future Contract Calculator

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
from pylab import rcParams
rcParams['figure.figsize'] = (12,8)
```

```python
# USD M Future

# collateral = input('Please Enter Collateral Amount: ')
# leverage_factor = input('Please Enter Leverage Factor: ')
# underlying_entry_price = input('Please Enter Underlying Entry Price: ')
# contract_size = input('Please Enter Contract Size: ')
collateral = 4000
leverage_factor = 20
underlying_entry_price = 22000
contract_size = 0.5
margin = (underlying_entry_price*contract_size)/leverage_factor
balance = collateral - margin
exposure = underlying_entry_price * contract_size
price_delta = balance/contract_size
long_liquidation_price = underlying_entry_price-price_delta
short_liquidation_price = underlying_entry_price+price_delta

print(f'USD-M Future Contracts Summary')
# print('\n')
print(f'Collateral U.S. Dollar for the Contract = ${collateral}')
print(f'Leverage Factor to Enter the Contract = {leverage_factor}x')
print(f'Underlying Entry Price = ${underlying_entry_price}')
print(f'Contract size = {contract_size} count')
# print('\n')
print(f'Margin for Entering the Contract = ${margin}')
print(f'Collateral Balance = ${balance}')
print(f'Risk Exposure = ${exposure}')
# print('\n')
print(f'Price Movement Tolerance = ${price_delta}')
print(f'Long Contract Liquidation Price = ${long_liquidation_price}')
print(f'Short Contract Liquidation Price = ${short_liquidation_price}')
plt.title('Long/Short Future Contracts Liquidation Price');
plt.xlabel('')
plt.ylabel('Price in USD')
plt.axhline(short_liquidation_price, c = 'red');
plt.axhline(underlying_entry_price, c = 'blue');
plt.axhline(long_liquidation_price, c = 'green');
plt.legend(["Short Liquidation Price ($)", "Contract Entry Price ($)", "Long Liquidation Price ($)"], frameon = False, loc='lower right', ncol=3);
```
